```python
import jieba
from gensim import corpora,models,similarities
import numpy as np
```

```python
import re
pattern = re.compile("[^\u4e00-\u9fa5^a-z^A-Z^0-9]")
```

```python
def cut_stop(dcut):
    stopwords = open('最全中文停用词.txt',encoding='utf-8')  # 这里加载停用词的路径
    words = ""
    for word in dcut:  # for循环遍历分词后的每个词语
        if word not in stopwords:  # 判断分词后的词语是否在停用词表内
            if word != '\t':
                words += word
                words += "/"
            #print('\n【去除停用词后的分词：】' + '\n' + words)
        content3 = words.replace('/', '')  # 去掉文本中的斜线
    return content3
```

```python
all_doc=[]
for i in range(1,3123):
    with open('岗位要求语料/require'+str(i)+'.txt',encoding='utf-8') as f:
        doc=f.read().replace(' ','')
        doc_cut= re.sub(pattern, '', doc)
        doc_n=cut_stop(doc_cut)
        all_doc.append(doc_n)
```

```python
all_doc
```

```python
all_doc_list = []
for doc in all_doc:
    doc_list = [word for word in jieba.cut(doc)]
    all_doc_list.append(doc_list)
```

```python
print(all_doc_list[0])
```

```python
with open('课程.txt',encoding='utf-8') as f:
    doc_test=f.read()
```

```python
doc_test
```

```python
doc_test= re.sub(pattern, '', doc_test)
doc_test
```

```python
doc_test_list = [word for word in jieba.cut(doc_test)]
doc_test_list
```

```python
dictionary = corpora.Dictionary(all_doc_list)
```

```python
dictionary.keys()
```

```python
dictionary.token2id
```

```python
corpus = [dictionary.doc2bow(doc) for doc in all_doc_list]
corpus
```

```python
doc_test_vec = dictionary.doc2bow(doc_test_list)
doc_test_vec
```

```python
tfidf = models.TfidfModel(corpus)
```

```python
tfidf[doc_test_vec]
```

```python
index = similarities.SparseMatrixSimilarity(tfidf[corpus], num_features=len(dictionary.keys()))
sim = index[tfidf[doc_test_vec]]
sim
```
